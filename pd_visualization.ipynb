```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

***Series***

```python
data = pd.Series(np.random.randn(1000), index=np.arange(1000))
print(data)
```

```python
data = data.cumsum()
data.plot()
plt.show()
```

***DataFrame***

```python
data1 = pd.DataFrame(np.random.randn(1000,4), index=np.arange(1000), columns=list('ABCD'))
print(data1.head())
```

```python
data1 = data1.cumsum()
data1.plot()
plt.show()  #plot methods: bar hist box kde area scatter hexbin pie
```

```python
ax = data1.plot.scatter(x='A', y='B', color='DarkBlue', label='Class 1')
data1.plot.scatter(x='A', y='C', color='DarkGreen', label='Class 2', ax = ax)
data1.plot.scatter(x='A', y='D', color='DarkRed', label='Class 3', ax = ax)
plt.show
```
